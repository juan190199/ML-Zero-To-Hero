```python
from sklearn import datasets

from unsupervised_learning.k_means import KMeans
from utils.misc import Plot
```

```python
# Generate data
X, y = datasets.make_blobs()

clf = KMeans(k=3)
y_pred = clf.predict(X)
```

```python
p = Plot()
p.plot_in_2d(X, y_pred, title="K-Means clustering")
p.plot_in_2d(X, y, title="True clustering")
```
